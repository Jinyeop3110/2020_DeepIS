from datas.preprocess import random_crop3d, elastic_transform, random_flip, random_rotate

```python
from glob import glob
data_path = "/data/00_Nuclues_segmentation/00_data/2D/New(50_Cells)/Balance/Train"
npys = glob(data_path + "/*.npy")

from random import shuffle
shuffle(list(npys))

len(npys)
```

```python
import numpy as np
import matplotlib.pyplot as plt

for npy in npys[:10]:
    npy_img = np.load(npy)
    img, mask = npy_img[:, :448], npy_img[:, 448:]
    imgs = [img, mask]
    elastic = elastic_transform(*imgs)
    crop = random_crop2d(*imgs)
    flip = random_flip2d(*imgs)
    rotate = random_rotate2d(*imgs)
    
    plt.figure(figsize=(15,15))
    plt.subplot(151); plt.title("origin")
    plt.imshow(img, cmap="gray")
    plt.subplot(152); plt.title("elastic")
    plt.imshow(elastic[0], cmap="gray")
    plt.subplot(153); plt.title("crop")
    plt.imshow(crop[0], cmap="gray")
    plt.subplot(154); plt.title("flip")
    plt.imshow(flip[0], cmap="gray")
    plt.subplot(155); plt.title("rotate")
    plt.imshow(rotate[0], cmap="gray")
```
